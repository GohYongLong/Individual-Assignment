```python
# Importing Libraries

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sentence_transformers import SentenceTransformer
```

```python
# Selecting Dataset from Drive

file_path = "/content/drive/MyDrive/Colab Notebooks/Individual Assignment/Individual 1 choice A spam (Text).csv"
df = pd.read_csv(file_path)

# Check the first row
print(df.head())

# Defininf X and Y
X=df['Message']
Y=df['Category']

# Checking for Class Imbalance
class_counts = df['Category'].value_counts()
print("Original Dataset Class Distribution:\n", class_counts)

# Show in %
class_percentage = df['Category'].value_counts(normalize=True) * 100
print("\nClass Distribution in Percentages:\n", class_percentage)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
X_CountVec = CountVectorizer().fit_transform(X)

X_CountVec
```

```python
# CountVectorizer Logistic Regression (No Stratified)

count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(X)

# Spliiting into train - test set
X_train, X_test, Y_train, Y_test = train_test_split(X_CountVec, Y)

# Training Logistic Model
model = LogisticRegression(random_state=69)
model.fit(X_train, Y_train)

# Prediction on test set
pred = model.predict(X_test)

# Confusion Matrix and Accuract
cm=confusion_matrix(Y_test, pred)
acc=accuracy_score(Y_test, pred)
print('Accuracy with simple CountVectorizer by SKLearn is ', acc)

```

```python
# CountVectorizer Random Forest (No Stratified)
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(X)

# Splitting into train-test set
X_train, X_test, Y_train, Y_test = train_test_split(X_CountVec, Y)

# Training Random Forest Model
model = RandomForestClassifier(n_estimators=100, random_state=69)
model.fit(X_train, Y_train)

# Prediction on test set
pred = model.predict(X_test)

# Confusion Matrix and Accuracy
cm = confusion_matrix(Y_test, pred)
acc = accuracy_score(Y_test, pred)
print('Accuracy with simple CountVectorizer using Random Forest is ', acc)
```

```python
# CountVectorizer Logistic Regression (With Stratified)

count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(X)

# Spliiting into train - test set
X_train, X_test, Y_train, Y_test = train_test_split(X_CountVec, Y, test_size=0.2, random_state=69, stratify=Y)

# Training Logistic Model
model = LogisticRegression()
model.fit(X_train, Y_train)

# Prediction on test set
pred = model.predict(X_test)

# Confusion Matrix and Accuract
cm=confusion_matrix(Y_test, pred)
acc=accuracy_score(Y_test, pred)
print('Accuracy with simple CountVectorizer by SKLearn is ', acc)
```

```python
# CountVectorizer Random Forest (With Stratified)
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(X)

# Splitting into train-test set
X_train, X_test, Y_train, Y_test = train_test_split(X_CountVec, Y, test_size=0.2, random_state=69, stratify=Y)

# Training Random Forest Model
model = RandomForestClassifier(n_estimators=100, random_state=69)
model.fit(X_train, Y_train)

# Prediction on test set
pred = model.predict(X_test)

# Confusion Matrix and Accuracy
cm = confusion_matrix(Y_test, pred)
acc = accuracy_score(Y_test, pred)
print('Accuracy with simple CountVectorizer using Random Forest is ', acc)
```

```python
# TDIF Logistic Regression (With Stratified)

# Convert text to CountVectorizer representation
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(X)

# Splitting into train-test set
X_train, X_test, Y_train, Y_test = train_test_split(X_count, Y, test_size=0.2, random_state=69, stratify=Y
)

# Fit and transform training data, transform test data
tfidf_transformer = TfidfTransformer()
X_train = tfidf_transformer.fit_transform(X_train)
X_test = tfidf_transformer.transform(X_test)

# Initialize and train Logistic Regression model on TF-IDF features
model = LogisticRegression()
model.fit(X_train, Y_train)

# Make predictions on TF-IDF transformed test data
pred = model.predict(X_test)

# Calculate Confusion Matrix and Accuracy
cm = confusion_matrix(Y_test, pred)
acc = accuracy_score(Y_test, pred)
print('Accuracy with TF-IDF using Logistic Regression is ', acc)
```

```python
# TDIF Random Forest (With Stratified)

# Convert text to CountVectorizer representation
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(X)

# Splitting into train-test set
X_train, X_test, Y_train, Y_test = train_test_split(X_count, Y, test_size=0.2, random_state=69, stratify=Y)

# Fit and transform training data, transform test data
tfidf_transformer = TfidfTransformer()
X_train = tfidf_transformer.fit_transform(X_train)
X_test = tfidf_transformer.transform(X_test)

# Initialize and train Random Forest model on TF-IDF features
model = RandomForestClassifier(n_estimators=100, random_state=69)
model.fit(X_train, Y_train)

# Make predictions on TF-IDF transformed test data
pred = model.predict(X_test)

# Calculate Confusion Matrix and Accuracy
cm = confusion_matrix(Y_test, pred)
acc = accuracy_score(Y_test, pred)
print('Accuracy with TF-IDF using Random Forest is ', acc)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
X_emb = model.encode(X)
X_emb.shape
```

```python
# BERT Logistic Regression (With Stratified)

# Splitting into train-test set
X_train, X_test, Y_train, Y_test = train_test_split(X_emb, Y, test_size=0.2, random_state=69, stratify=Y)

# Initialize and train Logistic Regression model on BERT embeddings
model = LogisticRegression()
model.fit(X_train, Y_train)

# Make predictions on BERT transformed test data
pred = model.predict(X_test)

# Calculate Confusion Matrix and Accuracy
cm = confusion_matrix(Y_test, pred)
acc = accuracy_score(Y_test, pred)
print('Accuracy using BERT with Logistic Regression is ', acc)
```

```python
# BERT Random Forest (With Stratified)

# Splitting into train-test set
X_train, X_test, Y_train, Y_test = train_test_split(X_emb, Y, test_size=0.2, random_state=69, stratify=Y)

# Initialize and train Random Forest model on BERT embeddings
model = RandomForestClassifier(n_estimators=100, random_state=69)
model.fit(X_train, Y_train)

# Make predictions on BERT transformed test data
pred = model.predict(X_test)

# Calculate Confusion Matrix and Accuracy
cm = confusion_matrix(Y_test, pred)
acc = accuracy_score(Y_test, pred)
print('Accuracy using BERT with Random Forest is ', acc)
```
