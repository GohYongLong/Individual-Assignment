```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive//Colab Notebooks/Individual Assignment/Individual 1 choice A spam (Text).csv")

# Check the first row
print(df.head())

X=df['Message']
Y=df['Category']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
X_CountVec = CountVectorizer().fit_transform(X)
```

```python
X_CountVec
```

```python
from sklearn.model_selection import train_test_split

# Spliiting into train - test set
X_train, X_test, Y_train, Y_test = train_test_split(X_CountVec, Y)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# Training Logistic Model
model = LogisticRegression()
model.fit(X_train, Y_train)

# Prediction on test set
pred = model.predict(X_test)

# Confusion Matrix and Accuract
cm=confusion_matrix(Y_test, pred)
acc=accuracy_score(Y_test, pred)
print('Accuracy with simple CountVectorizer by SKLearn is ', acc)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

# Fit and transform training data, transform test data
tfidf_transformer = TfidfTransformer()
X_train = tfidf_transformer.fit_transform(X_train)
X_test = tfidf_transformer.transform(X_test)

# Initialize and train Logistic Regression model on TF-IDF features
model = LogisticRegression()
model.fit(X_train, Y_train)

# Make predictions on TF-IDF transformed test data
pred = model.predict(X_test)

# Calculate Confusion Matrix and Accuracy
cm=confusion_matrix(Y_test, pred)
acc=accuracy_score(Y_test, pred)
print('Accuracy with TFIDF is ',acc)
```
